```python
df_scope = pd.read_pickle('data/LDA_Messages_persisted_with_language.pkl')
```

```python
dataset = 'Udbetalingsteam'
initiatedby = 'Initiatedby_AKA'
n_gram = 3
sample_size= 10000
no_words= 5000
no_above = 0.001
no_below= 2 # filter out tokens that appear in less than 15 documents
random_state=1
research_scope = 'Udbetalingteam'
lda_num_topics = 50
lda_chunksize = 1000
lda_passes = 50
```

```python
df_test = df_scope[(df_scope['FirstMessage'].isnull()==False) & (df_scope['ThreadResponsibleDepartmentTeam'].str.contains(dataset)==True)].copy(deep=True)
df_test.shape
```

```python
dictionary = PrepareDictionary(df_test, 'stemmed_text',no_above,no_below)
```

```python
corpus = df_test.bow
```

```python
%%time
from gensim.models.ldamulticore import LdaMulticore
LDAmodel_scope = LdaMulticore(corpus=corpus,#mm,
                        id2word=dictionary,
                        num_topics=lda_num_topics,
                        workers=4,
                        chunksize=lda_chunksize,
                        passes=lda_passes,
                        alpha='asymmetric',
                        random_state=random_state)
```

```python
topic_distribution_barplot(LDAmodel_scope, df_test)
```
